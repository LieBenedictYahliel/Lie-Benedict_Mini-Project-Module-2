## Latar Belakang : 

Dalam era digital saat ini, konsumen semakin aktif dalam memberikan penilaian dan ulasan terhadap produk yang mereka beli secara daring. Penilaian tersebut, yang tercermin dalam rating produk, menjadi indikator penting dalam mengukur kepuasan dan persepsi konsumen terhadap kualitas suatu barang. Oleh karena itu, memahami faktor-faktor yang memengaruhi rating produk menjadi hal yang krusial, terutama bagi pelaku bisnis dan pemilik merek.

Beberapa faktor yang diduga memengaruhi penilaian konsumen antara lain harga produk, besarnya diskon, dan brand. Secara umum, ada anggapan bahwa harga yang lebih tinggi mencerminkan kualitas yang lebih baik, sehingga akan mendapatkan penilaian yang lebih tinggi pula. Namun, realitas di lapangan tidak selalu sejalan. Sebaliknya, produk dengan harga murah dan diskon besar belum tentu mendapatkan rating rendah sebagian bahkan justru menjadi produk favorit konsumen.

Selain itu, nama dan reputasi brand juga memainkan peran penting. Brand yang dikenal akan kualitas dan kepuasan pelanggan cenderung memiliki rating lebih tinggi dibandingkan brand baru atau yang memiliki masalah dalam layanan maupun produk.

```python
import requests 
import pandas as pd
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
dummy_API = "https://dummyjson.com/products"
```

```python
# Mengambil data
resultRequest = requests.get(dummy_API)
resultRequest
```

```python
# Menampilkan data
resultRequest.json()
```

```python
# Menampilkan data
resultRequest.json()['products']
```

```python
# Membuat list kosong
dataId = []
title = []
description = []
category = []
price = []
discount = []
rating = []
stock = []
brand = []
sku = []
weight = []
dimensions = []
warrantyInformation = []
shippingInformation = []
availabilityStatus = []
reviews = []

# Memasukkan data ke dalam list
for data in resultRequest.json()['products']:
    dataId.append(data.get('id'))
    title.append(data.get('title'))
    description.append(data.get('description'))
    category.append(data.get('category'))
    price.append(data.get('price'))
    discount.append(data.get('discountPercentage'))
    rating.append(data.get('rating'))
    stock.append(data.get('stock'))
    brand.append(data.get('brand')) 
    sku.append(data.get('sku'))
    weight.append(data.get('weight'))
    dimensions.append(data.get('dimensions'))
    warrantyInformation.append(data.get('warrantyInformation'))
    shippingInformation.append(data.get('shippingInformation'))
    availabilityStatus.append(data.get('availabilityStatus'))
    reviews.append(data.get('reviews'))
    
```

```python
# Membuat DataFrame
data = {
    'id': dataId,
    'title': title,
    'description': description,
    'category': category,
    'price': price,
    'discount': discount,
    'rating': rating,
    'stock': stock,
    'brand': brand,
    'sku': sku,
    'weight': weight,
    'dimensions': dimensions,
    'warrantyInformation': warrantyInformation,
    'shippingInformation': shippingInformation,
    'availabilityStatus': availabilityStatus,
    'reviews': reviews
}
df = pd.DataFrame(data)
df
```

```python
# Isi semua NaN (termasuk None sebelumnya) dengan 'Other'
df['brand'].fillna('Other', inplace=True)
df
```

```python
df['category'] = df['category'].transform(lambda x: x.strip().title())
df
```

```python
# Mengganti nama kolom
df.rename(columns={'title': 'product_name'}, inplace=True) 
```

```python
# Menghapus beberapa kolom
df.drop(['id', 'description', 'sku', 'weight', 'dimensions', 'warrantyInformation', 'shippingInformation', 'availabilityStatus', 'reviews'], axis=1, inplace=True, errors= 'ignore')
df
```

# Analisa data

### 1. Apakah produk dengan diskon tinggi cenderung mendapatkan rating lebih rendah karena ekspektasi pelanggan?

```python
# Menghitung korelasi antara harga (price), diskon (discount), dan rating
correlation = df[['price', 'discount', 'rating']].corr()
print(correlation['rating'].round(2).sort_values(ascending=False)) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ukuran plot
plt.figure(figsize=(14, 6))

# 1. Rating vs Price dengan garis tren
plt.subplot(1, 2, 1)
sns.regplot(data=df, x='price', y='rating', scatter_kws={'alpha':0.6})
plt.title(f'Rating vs Price\nCorrelation: {correlation["rating"]["price"]:.2f}')
plt.xlabel('Price')
plt.ylabel('Rating')

# 2. Rating vs Discount dengan garis tren
plt.subplot(1, 2, 2)
sns.regplot(data=df, x='discount', y='rating', scatter_kws={'alpha':0.6}, color='orange')
plt.title(f'Rating vs Discount\nCorrelation: {correlation["rating"]["discount"]:.2f}')
plt.xlabel('Discount')
plt.ylabel('Rating')

plt.tight_layout()
plt.show()
```

### insight : 
Harga (Price) punya korelasi positif lemah terhadap rating (+0.16):

Artinya, semakin mahal suatu produk, cenderung memiliki rating sedikit lebih tinggi.

Meskipun korelasinya lemah, ini bisa menunjukkan bahwa pelanggan mungkin menilai produk mahal sebagai "lebih berkualitas".

📉 Diskon (Discount) punya korelasi negatif lemah terhadap rating (−0.20):

Semakin besar diskon, rating cenderung sedikit lebih rendah.

Mungkin produk yang sering diskon dianggap kurang eksklusif, bermasalah stok, atau kurang laku, sehingga persepsi pelanggan juga rendah.

Bisa juga karena konsumen punya ekspektasi lebih tinggi saat beli dengan diskon, lalu kecewa.

### 2. Brand mana yang memiliki rata-rata rating tertinggi berdasarkan ulasan pelanggan?

```python
# Menghitung rata-rata rating untuk setiap brand
rating_per_brand=df.groupby('brand')['rating'].mean().round(2).sort_values(ascending=False).reset_index()
rating_per_brand
```

```python
bar_rating_brand = px.bar(
    rating_per_brand,
    x="brand",
    y="rating",
    title="Rata-Rata Rating per Brand",
    text="rating",
    color="brand",
    labels={
        "brand": "Merek",
        "rating": "Rata-Rata Rating"
    }
)
bar_rating_brand.update_layout(
    title_x=0.5,
    title_xanchor="center"
)
bar_rating_brand.show()
```

### insight : 
Knoll (4.88), Velvet Touch (4.64), dan FreshCo (4.46) menduduki peringkat teratas.

Ini menunjukkan bahwa konsumen sangat puas terhadap produk dari brand ini.

Essence (2.56) dan Glamour Beauty (2.86) memiliki rating terendah.

Perlu dilakukan evaluasi: apakah masalahnya di produk, harga, pelayanan, atau ekspektasi pelanggan yang tidak terpenuhi?

### 3. Produk apa dengan rating tertinggi tetapi harga paling murah?

```python
produk_terbaik = df.sort_values(by=['rating', 'price'], ascending=[False, True]).head(1)
produk_terbaik
```

```python
import matplotlib.pyplot as plt

# Urutkan data berdasarkan rating
df_sorted = df.sort_values(by='rating', ascending=True)

# Ukuran figure disesuaikan dengan jumlah data
plt.figure(figsize=(10, len(df_sorted) * 0.4))

# Buat horizontal bar chart berdasarkan rating
plt.barh(df_sorted['product_name'], df_sorted['rating'], color='skyblue')
plt.xlabel('Rating')
plt.title('Perbandingan Rating Semua Produk (dengan Harga di Label)')

# Tampilkan HARGA di samping bar, bukan rating
for index, (rating, price) in enumerate(zip(df_sorted['rating'], df_sorted['price'])):
    plt.text(rating + 0.05, index, f"${price:.2f}", va='center')  # Menampilkan harga

plt.tight_layout()
plt.show()
```

## insight : 

dengan rating tertinggi (4.93), harga termurah (2.49), dan stok tinggi (99), artinya:

Produk dengan rating tinggi tidak selalu mahal. Bahkan, dalam kasus ini, produk termurah justru paling disukai pelanggan.

Ini membuka kemungkinan bahwa harga terjangkau dapat mendorong rating positif, terutama untuk kebutuhan sehari-hari seperti bahan makanan.

### 4.Apakah kategori produk tertentu cenderung memiliki stok lebih rendah karena tingginya permintaan?



```python

# Hitung stok rata-rata per kategori
stok_per_kategori = df.groupby('category')['stock'].mean().sort_values()

# Visualisasi dalam bentuk barplot
plt.figure(figsize=(10,6))
sns.barplot(x=stok_per_kategori.values, y=stok_per_kategori.index, palette='viridis')
plt.xlabel('Rata-rata Jumlah Stok')
plt.ylabel('Kategori Produk')
plt.title('Rata-rata Stok Produk per Kategori')
plt.tight_layout()
plt.show()
```

### Insight :
Kategori groceries memiliki rata-rata stok terendah, sementara beauty memiliki stok tertinggi. dapat disimpulkan bahwa permintaan terhadap groceries lebih tinggi karena sifatnya yang esensial dan cepat habis. Sebaliknya, produk beauty cenderung bergerak lambat karena siklus pembelian yang lebih jarang. Namun, stok rendah belum tentu berarti permintaan tinggi, bisa juga karena strategi penyimpanan.
